```python
import sqlalchemy
from sqlalchemy import Table, Column, Integer, String, MetaData, ForeignKey, create_engine

engine = create_engine('sqlite:///Acup_last_11.db')
meta = MetaData() # MetaData is a container object that keeps together many different features of a database (or multiple databases) being described.
Symptom = Table('Symptom', meta,
             Column('Symp_id', String, primary_key = True),
             Column('Symp_name', String))

Value = Table('Value', meta,
              Column('Value_id', String, primary_key = True),
              Column('Keyword', String),
              Column('Symp_id', ForeignKey('Symptom.Symp_id')))

Acup = Table('Acup', meta,
             Column('Acup_id', String, primary_key = True),
             Column('Acup_name', String),
             Column('Acup_position', String))

Food = Table('Food', meta,
             Column('Food_id', String, primary_key = True),
             Column('Food_name', String),
             Column('How_to_eat', String),
             Column('Symp_id', ForeignKey('Symptom.Symp_id')))

Acup_Method = Table('Acup_Method', meta,
               Column('Acup_Method_id', String, primary_key = True),
               Column('Acup_id', ForeignKey('Acup.Acup_id')),
               Column('Symp_id', ForeignKey('Symptom.Symp_id')))
meta.create_all(engine)
```
